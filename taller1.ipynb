**1. Ejercicio de tokens vs precio**

```python
import numpy as np
```

```python
tokens = list(range(1_000_000))
```

```python
print(tokens[:10])
print(tokens[-5:])
```

```python

precio = []

for t in tokens:

    if t <= 300_000:
        precio.append((8.16 / 300_000) * t)

    else:
        precio.append(
            8.16 + ((45 - 8.16) / (1_000_000 - 300_000)) * (t - 300_000)
        )
```

```python
import matplotlib.pyplot as plt

plt.plot(tokens, precio)

plt.axvline(x=300_000, linestyle='--', color='red', label='Cambio contexto')

plt.xlabel("Tokens")
plt.ylabel("Precio")
plt.title("Precio vs Tokens")

plt.grid(True)
plt.show()
```

**2. Ejercicio de la biseccion**

Encontrar todas las raíces de la ecuación f(x)=1/4(x3+3x2−6x−8)
 usando el método de la bisección. Grafique. Modifique el número de iteraciones a 5. Pruebe los siguientes rangos de búsqueda ([-5 3], [-4.7 2.5], [-4.5 2.7], [-3 -2]). 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import bisect


# Funcion del ejercicio

def f(x):
    return (1/4) * (x**3 + 3*x**2 - 6*x - 8)


# Metodo de Biseccion

def biseccion_iter(a, b, n_iter=5):
    print(f"\nBISECCIÓN en intervalo [{a}, {b}]")

    if f(a) * f(b) >= 0:
        print("No hay cambio de signo en este intervalo.")
        return None

    for i in range(1, n_iter + 1):
        c = (a + b) / 2
        fc = f(c)

        print(f"Iteración {i}:")
        print(f"   a = {a:.6f}, b = {b:.6f}")
        print(f"   c = {c:.6f}, f(c) = {fc:.6f}")

        if f(a) * fc < 0:
            b = c
        else:
            a = c

    return (a + b) / 2

#Intervalos que vamos a evaluar
intervalos = [
    (-5, 3),
    (-4.7, 2.5),
    (-4.5, 2.7),
    (-3, -2)
]


raices = []


print(" MÉTODO DE BISECCIÓN")


for a, b in intervalos:
    raiz = biseccion_iter(a, b, 5)
    if raiz is not None:
        raices.append(raiz)
        print(f"Raíz aproximada: {raiz:.6f}")

#Grafica

x = np.linspace(-6, 4, 400)
y = f(x)

plt.figure(figsize=(9,5))
plt.plot(x, y, label="f(x)", color="blue")
plt.axhline(0, color="black", linewidth=1)

# raíces
for r in raices:
    plt.plot(r, f(r), 'ro')
    plt.text(r, f(r), f"{r:.2f}")

plt.title("Método de Bisección")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
