# 2. Matrices no singulares e inversas

Una matriz es **no singular** cuando su determinante es diferente de cero:

\[
\det(A)\neq 0
\]

En ese caso, la matriz posee inversa. Si

\[
\det(A)=0,
\]

la matriz es singular y no tiene inversa.

---

## a.

Sea

\[
A=
\begin{bmatrix}
4 & 2 & 6\\
3 & 0 & 7\\
-2 & -1 & -3
\end{bmatrix}
\]

Calculamos su determinante mediante expansión por la primera fila:

\[
\det(A)=
4
\begin{vmatrix}
0 & 7\\
-1 & -3
\end{vmatrix}
-
2
\begin{vmatrix}
3 & 7\\
-2 & -3
\end{vmatrix}
+
6
\begin{vmatrix}
3 & 0\\
-2 & -1
\end{vmatrix}
\]

\[
\det(A)
=
4[0(-3)-7(-1)]
-
2[3(-3)-7(-2)]
+
6[3(-1)-0(-2)]
\]

\[
\det(A)
=
4(7)-2(-9+14)+6(-3)
\]

\[
\det(A)=28-10-18
\]

\[
\boxed{\det(A)=0}
\]

También se puede observar que la tercera fila es igual a la primera fila multiplicada por \(-\frac12\):

\[
F_3=-\frac12F_1
\]

Por lo tanto, sus filas son linealmente dependientes.

### Conclusión

La matriz es **singular** y no posee inversa:

\[
\boxed{A^{-1}\text{ no existe}}
\]

# 4. Valores de \(\alpha\) para que la matriz sea singular

Sea la matriz:

\[
A=
\begin{bmatrix}
1 & -1 & \alpha\\
2 & 2 & 1\\
0 & \alpha & -\frac{3}{2}
\end{bmatrix}
\]

Una matriz es **singular** cuando su determinante es igual a cero:

\[
\det(A)=0
\]

Calculamos el determinante expandiendo por la primera fila:

\[
\det(A)=
1
\begin{vmatrix}
2 & 1\\
\alpha & -\frac{3}{2}
\end{vmatrix}
-
(-1)
\begin{vmatrix}
2 & 1\\
0 & -\frac{3}{2}
\end{vmatrix}
+
\alpha
\begin{vmatrix}
2 & 2\\
0 & \alpha
\end{vmatrix}
\]

Calculamos cada determinante de orden \(2\times 2\):

\[
\det(A)
=
1\left[
2\left(-\frac{3}{2}\right)-1(\alpha)
\right]
+
\left[
2\left(-\frac{3}{2}\right)-1(0)
\right]
+
\alpha\left[2(\alpha)-2(0)\right]
\]

Simplificamos:

\[
\det(A)=(-3-\alpha)+(-3)+2\alpha^2
\]

\[
\det(A)=2\alpha^2-\alpha-6
\]

Para que la matriz sea singular:

\[
2\alpha^2-\alpha-6=0
\]

Factorizamos:

\[
2\alpha^2-\alpha-6
=
(2\alpha+3)(\alpha-2)
\]

Entonces:

\[
(2\alpha+3)(\alpha-2)=0
\]

Igualamos cada factor a cero:

\[
2\alpha+3=0
\]

\[
2\alpha=-3
\]

\[
\alpha=-\frac{3}{2}
\]

O bien:

\[
\alpha-2=0
\]

\[
\alpha=2
\]

Por lo tanto, los valores que hacen que la matriz sea singular son:

\[
\boxed{\alpha=2 \quad \text{o} \quad \alpha=-\frac{3}{2}}
\]

# 5. Resolución de sistemas lineales

Los sistemas están escritos como:

\[
LU\mathbf{x}=\mathbf{b}
\]

Para resolverlos, se utiliza una variable auxiliar:

\[
U\mathbf{x}=\mathbf{y}
\]

De esta manera, primero resolvemos:

\[
L\mathbf{y}=\mathbf{b}
\]

y después:

\[
U\mathbf{x}=\mathbf{y}
\]

---

\[

-1 & 0 & 1
\end{bmatrix}
\begin{bmatrix}

0 & 0 & 3
\end{bmatrix}
\begin{bmatrix}
x_1\\
x_2\\
x_3
\end{bmatrix}
=
\begin{bmatrix}
2\\
-1\\
1
\end{bmatrix}
\]

\[

\begin{bmatrix}

\end{bmatrix},

\begin{bmatrix}
2 & 3 & -1\\

0 & 0 & 3
\end{bmatrix}
\]

# 6. Descomposición \(LU\)

Se desea factorizar cada matriz de la forma:

\[
A=LU
\]

donde:

- \(L\) es una matriz triangular inferior.
- \(U\) es una matriz triangular superior.
- Los elementos de la diagonal principal de \(L\) cumplen:

\[
l_{ii}=1
\]

Este procedimiento corresponde al método de **Doolittle**.

# 7. Resolución de sistemas mediante descomposición \(LU\)

Para resolver un sistema lineal

\[
A\mathbf{x}=\mathbf{b},
\]

se factoriza la matriz de coeficientes como:

\[
A=LU,
\]

donde:

- \(L\) es triangular inferior y tiene unos en su diagonal principal.
- \(U\) es triangular superior.

Entonces:

\[
LU\mathbf{x}=\mathbf{b}.
\]

Definimos un vector auxiliar:

\[
U\mathbf{x}=\mathbf{y}.
\]

De esta manera se resuelven dos sistemas:

\[
L\mathbf{y}=\mathbf{b}
\]

mediante sustitución hacia adelante, y luego:

\[
U\mathbf{x}=\mathbf{y}
\]

mediante sustitución hacia atrás.

---

## Algoritmo modificado

```text
Entrada: matriz A y vector b

1. Factorizar A como A = LU.

2. Resolver Ly = b mediante sustitución hacia adelante:

   Para i = 1 hasta n:
       y[i] = b[i]

       Para j = 1 hasta i - 1:
           y[i] = y[i] - L[i,j]y[j]

       y[i] = y[i] / L[i,i]

3. Resolver Ux = y mediante sustitución hacia atrás:

   Para i = n hasta 1:
       x[i] = y[i]

       Para j = i + 1 hasta n:
           x[i] = x[i] - U[i,j]x[j]

       x[i] = x[i] / U[i,i]

4. Mostrar x.
