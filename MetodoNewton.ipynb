```python
import numpy as np

xs = list(range(-2, 6))
xs = list(np.linspace(-1, 5, 100))
```

```python
def f(x):
    _y = x**3 - 3 * x**2 + x - 1
    if np.isscalar(x):
        print(f"{x:.2f} \t {_y:.2f}")
    return _y


def df(x):
    _dy = 3 * x**2 - 6 * x + 1

    return _dy
```

```python
ys = [f(x) for x in xs]
```

```python
import matplotlib.pyplot as plt


plt.plot(xs, ys,)
plt.xlabel("xs")
plt.ylabel("ys")
plt.title("Gráfica de ys vs xs")
plt.grid(True)
plt.show()


```

```python
import matplotlib.pyplot as plt

# -------- Primer caso: x0 = -10 --------
x_vals1 = []
x = -10

for i in range(20):
    x_vals1.append(x)
    x = x - f(x)/df(x)

iters1 = list(range(len(x_vals1)))


# -------- Segundo caso: x0 = -8 --------
x_vals2 = []
x = -8

for i in range(20):
    x_vals2.append(x)
    x = x - f(x)/df(x)

iters2 = list(range(len(x_vals2)))


plt.figure(figsize=(18,6))

plt.plot(iters1, x_vals1, marker='o', label='x0 = -10')
plt.plot(iters2, x_vals2, marker='s', label='x0 = -8')

# Etiquetas de puntos primer caso
for i, x in enumerate(x_vals1):
    plt.text(i, x, f"{x:.2f}", fontsize=8)

# Etiquetas de puntos segundo caso
for i, x in enumerate(x_vals2):
    plt.text(i, x, f"{x:.2f}", fontsize=8)

plt.title("Comparación Método de Newton")
plt.xlabel("Iteración")
plt.ylabel("xₙ")
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# rango para dibujar tangentes
x_plot = np.linspace(-12, 5, 400)

# =========================
# Newton x0 = -10
# =========================
x_vals1 = []
x1 = -10

for i in range(10):
    x_vals1.append(x1)
    x1 = x1 - f(x1)/df(x1)

# =========================
# Newton x0 = -8
# =========================
x_vals2 = []
x2 = -8

for i in range(10):
    x_vals2.append(x2)
    x2 = x2 - f(x2)/df(x2)


# =========================
# GRÁFICA
# =========================
plt.figure(figsize=(18,6))

# función original
plt.plot(x_plot, f(x_plot), 'k', label="f(x)")

# -------- tangentes x0 = -10 --------
for x0 in x_vals1:
    y0 = f(x0)
    m = df(x0)
    tangent = y0 + m*(x_plot - x0)
    plt.plot(x_plot, tangent, '--', alpha=0.4, color='blue')

# -------- tangentes x0 = -8 --------
for x0 in x_vals2:
    y0 = f(x0)
    m = df(x0)
    tangent = y0 + m*(x_plot - x0)
    plt.plot(x_plot, tangent, '--', alpha=0.4, color='orange')

# puntos de iteración
plt.scatter(x_vals1, [f(x) for x in x_vals1], color='blue', label="x0 = -10")
plt.scatter(x_vals2, [f(x) for x in x_vals2], color='orange', label="x0 = -8")

plt.axhline(0, color='black', linewidth=1)
plt.title("Método de Newton con tangentes")
plt.legend()
plt.grid(True)

plt.show()
```
