1. Dados los datos:
xi 4.0 4.2 4.5 4.7 5.1 5.5 5.9 6.3 6.8 7.1
yi 102.56 130.11 113.18 142.05 167.53 195.14 224.87 256.73 299.50 326.72

a. Construya el polinomio por mínimos cuadrados de grado 1 y calcule el error.

b. Construya el polinomio por mínimos cuadrados de grado 2 y calcule el error.

c. Construya el polinomio por mínimos cuadrados de grado 3 y calcule el error.

d. Construya el polinomio por mínimos cuadrados de la forma 𝑏𝑒௔௫ y calcule el error.

e. Construya el polinomio por mínimos cuadrados de la forma 𝑏𝑥௔
 y calcule el error. 







```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# Datos
# ==========================

x = np.array([4.0,4.2,4.5,4.7,5.1,5.5,5.9,6.3,6.8,7.1])

y = np.array([102.56,
              130.11,
              113.18,
              142.05,
              167.53,
              195.14,
              224.87,
              256.73,
              299.50,
              326.72])

# ==========================
# Función para calcular error
# ==========================

def error(y_real, y_aprox):
    return np.sum((y_real-y_aprox)**2)

# =====================================================
# a) Polinomio grado 1
# =====================================================

coef1 = np.polyfit(x,y,1)

p1 = np.poly1d(coef1)

y1 = p1(x)

error1 = error(y,y1)

print("===================================")
print("POLINOMIO GRADO 1")
print(p1)
print("Error =",error1)

# =====================================================
# b) Polinomio grado 2
# =====================================================

coef2 = np.polyfit(x,y,2)

p2 = np.poly1d(coef2)

y2 = p2(x)

error2 = error(y,y2)

print("\n===================================")
print("POLINOMIO GRADO 2")
print(p2)
print("Error =",error2)

# =====================================================
# c) Polinomio grado 3
# =====================================================

coef3 = np.polyfit(x,y,3)

p3 = np.poly1d(coef3)

y3 = p3(x)

error3 = error(y,y3)

print("\n===================================")
print("POLINOMIO GRADO 3")
print(p3)
print("Error =",error3)

# =====================================================
# d) Modelo exponencial
# y = b*e^(ax)
# ln(y)=ln(b)+ax
# =====================================================

lny = np.log(y)

coef_exp = np.polyfit(x,lny,1)

a = coef_exp[0]
b = np.exp(coef_exp[1])

y_exp = b*np.exp(a*x)

error_exp = error(y,y_exp)

print("\n===================================")
print("MODELO EXPONENCIAL")
print(f"y = {b:.6f} * e^({a:.6f}x)")
print("Error =",error_exp)

# =====================================================
# e) Modelo potencial
# y = b*x^a
# ln(y)=ln(b)+a ln(x)
# =====================================================

lnx = np.log(x)

coef_pot = np.polyfit(lnx,lny,1)

a2 = coef_pot[0]
b2 = np.exp(coef_pot[1])

y_pot = b2*(x**a2)

error_pot = error(y,y_pot)

print("\n===================================")
print("MODELO POTENCIAL")
print(f"y = {b2:.6f} * x^({a2:.6f})")
print("Error =",error_pot)

# =====================================================
# Gráfica
# =====================================================

xp = np.linspace(min(x),max(x),500)

plt.figure(figsize=(10,6))

plt.scatter(x,y,
            color='black',
            s=60,
            label='Datos')

plt.plot(xp,p1(xp),
         linewidth=2,
         label='Grado 1')

plt.plot(xp,p2(xp),
         linewidth=2,
         label='Grado 2')

plt.plot(xp,p3(xp),
         linewidth=2,
         label='Grado 3')

plt.plot(xp,
         b*np.exp(a*xp),
         linewidth=2,
         label='Exponencial')

plt.plot(xp,
         b2*(xp**a2),
         linewidth=2,
         label='Potencial')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Ajustes por mínimos cuadrados")
plt.grid(True)
plt.legend()

plt.show()

# =====================================================
# Tabla de errores
# =====================================================

print("\n==============================")
print("RESUMEN DE ERRORES")
print("==============================")
print(f"Grado 1       : {error1:.4f}")
print(f"Grado 2       : {error2:.4f}")
print(f"Grado 3       : {error3:.4f}")
print(f"Exponencial   : {error_exp:.4f}")
print(f"Potencial     : {error_pot:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# Datos
# ==========================

x = np.array([0.2, 0.3, 0.6, 0.9, 1.1, 1.3, 1.4, 1.6])

y = np.array([0.050446,
              0.098426,
              0.33277,
              0.72660,
              1.0972,
              1.5697,
              1.8487,
              2.5015])

# ==========================
# Función para calcular el error
# ==========================

def error(y_real, y_aprox):
    return np.sum((y_real - y_aprox)**2)

# =====================================================
# a) Polinomio grado 1
# =====================================================

coef1 = np.polyfit(x, y, 1)
p1 = np.poly1d(coef1)
y1 = p1(x)
error1 = error(y, y1)

print("===================================")
print("POLINOMIO GRADO 1")
print(p1)
print("Error =", error1)

# =====================================================
# b) Polinomio grado 2
# =====================================================

coef2 = np.polyfit(x, y, 2)
p2 = np.poly1d(coef2)
y2 = p2(x)
error2 = error(y, y2)

print("\n===================================")
print("POLINOMIO GRADO 2")
print(p2)
print("Error =", error2)

# =====================================================
# c) Polinomio grado 3
# =====================================================

coef3 = np.polyfit(x, y, 3)
p3 = np.poly1d(coef3)
y3 = p3(x)
error3 = error(y, y3)

print("\n===================================")
print("POLINOMIO GRADO 3")
print(p3)
print("Error =", error3)

# =====================================================
# d) Modelo exponencial
# y = b*e^(ax)
# =====================================================

lny = np.log(y)

coef_exp = np.polyfit(x, lny, 1)

a = coef_exp[0]
b = np.exp(coef_exp[1])

y_exp = b * np.exp(a * x)

error_exp = error(y, y_exp)

print("\n===================================")
print("MODELO EXPONENCIAL")
print(f"y = {b:.6f} * e^({a:.6f}x)")
print("Error =", error_exp)

# =====================================================
# e) Modelo potencial
# y = b*x^a
# =====================================================

lnx = np.log(x)

coef_pot = np.polyfit(lnx, lny, 1)

a2 = coef_pot[0]
b2 = np.exp(coef_pot[1])

y_pot = b2 * (x ** a2)

error_pot = error(y, y_pot)

print("\n===================================")
print("MODELO POTENCIAL")
print(f"y = {b2:.6f} * x^({a2:.6f})")
print("Error =", error_pot)

# =====================================================
# Gráfica
# =====================================================

xp = np.linspace(min(x), max(x), 500)

plt.figure(figsize=(10,6))

plt.scatter(x, y, color='black', s=60, label='Datos')

plt.plot(xp, p1(xp), label='Grado 1', linewidth=2)
plt.plot(xp, p2(xp), label='Grado 2', linewidth=2)
plt.plot(xp, p3(xp), label='Grado 3', linewidth=2)
plt.plot(xp, b*np.exp(a*xp), label='Exponencial', linewidth=2)
plt.plot(xp, b2*(xp**a2), label='Potencial', linewidth=2)

plt.title("Ajustes por mínimos cuadrados")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.show()

# =====================================================
# Resumen de errores
# =====================================================

print("\n==============================")
print("RESUMEN DE ERRORES")
print("==============================")
print(f"Grado 1     : {error1:.6f}")
print(f"Grado 2     : {error2:.6f}")
print(f"Grado 3     : {error3:.6f}")
print(f"Exponencial : {error_exp:.6f}")
print(f"Potencial   : {error_pot:.6f}")
```

3. La siguiente tabla muestra los promedios de puntos del colegio de 20 especialistas en matemáticas y ciencias
computacionales, junto con las calificaciones que recibieron estos estudiantes en la parte de matemáticas de
la prueba ACT (Programa de Pruebas de Colegios Americanos) mientras estaban en secundaria. Grafique
estos datos y encuentre la ecuación de la recta por mínimos cuadrados para estos datos.



```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# Datos
# x = Puntuación ACT
# y = Promedio de puntos
# ==========================

x = np.array([
    28, 25, 28, 27, 28,
    33, 28, 29, 23, 27,
    29, 28, 27, 29, 21,
    28, 28, 26, 30, 24
])

y = np.array([
    3.84, 3.21, 3.23, 3.63, 3.75,
    3.20, 3.41, 3.38, 3.53, 2.03,
    3.75, 3.65, 3.87, 3.75, 1.66,
    3.12, 2.96, 2.92, 3.10, 2.81
])

# ==========================
# Recta por mínimos cuadrados
# ==========================

coef = np.polyfit(x, y, 1)

m = coef[0]
b = coef[1]

print(f"Pendiente (m) = {m:.6f}")
print(f"Intercepto (b) = {b:.6f}")

print(f"\nRecta de mínimos cuadrados:")
print(f"y = {m:.6f}x + {b:.6f}")

# ==========================
# Error cuadrático
# ==========================

y_aprox = m*x + b

SSE = np.sum((y - y_aprox)**2)

print(f"\nError SSE = {SSE:.6f}")

# ==========================
# Coeficiente de determinación
# ==========================

R2 = 1 - np.sum((y-y_aprox)**2)/np.sum((y-np.mean(y))**2)

print(f"R² = {R2:.6f}")

# ==========================
# Gráfica
# ==========================

xp = np.linspace(min(x)-1, max(x)+1, 200)

plt.figure(figsize=(8,6))

plt.scatter(x, y, color="blue", s=60, label="Datos")

plt.plot(xp,
         m*xp+b,
         color="red",
         linewidth=2,
         label="Recta de mínimos cuadrados")

plt.xlabel("Puntuación ACT")
plt.ylabel("Promedio de puntos")
plt.title("Regresión lineal por mínimos cuadrados")
plt.grid(True)
plt.legend()

plt.show()
```

4. El siguiente conjunto de datos, presentado al Subcomité Antimonopolio del Senado, muestra las
características comparativas de supervivencia durante un choque de automóviles de diferentes clases.
Encuentre la recta por mínimos cuadrados que aproxima estos datos (la tabla muestra el porcentaje de
vehículos que participaron en un accidente en los que la lesión más grave fue fatal o seria). 



```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# Datos
# x = Peso promedio (lb)
# y = Porcentaje de lesiones
# ==========================

x = np.array([4800, 3700, 3400, 2800, 1900])

y = np.array([3.1, 4.0, 5.2, 6.4, 9.6])

# ==========================
# Recta de mínimos cuadrados
# ==========================

coef = np.polyfit(x, y, 1)

m = coef[0]
b = coef[1]

print("Pendiente =", m)
print("Intercepto =", b)

print(f"\nRecta de mínimos cuadrados:")
print(f"y = {m:.8f}x + {b:.8f}")

# ==========================
# Valores aproximados
# ==========================

y_aprox = m*x + b

# Error cuadrático
SSE = np.sum((y - y_aprox)**2)

print(f"\nError SSE = {SSE:.6f}")

# Coeficiente de determinación
R2 = 1 - np.sum((y-y_aprox)**2)/np.sum((y-np.mean(y))**2)

print(f"R² = {R2:.6f}")

# ==========================
# Gráfica
# ==========================

xp = np.linspace(min(x)-100, max(x)+100, 300)

plt.figure(figsize=(8,6))

plt.scatter(x, y,
            color="blue",
            s=70,
            label="Datos")

plt.plot(xp,
         m*xp+b,
         color="red",
         linewidth=2,
         label="Recta de mínimos cuadrados")

plt.title("Peso promedio vs Porcentaje de lesiones")
plt.xlabel("Peso promedio (lb)")
plt.ylabel("Porcentaje (%)")
plt.grid(True)
plt.legend()

plt.show()
```
