```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# DATOS
# =========================
x = np.array([0, 0, 2, 2 ])
y = np.array([0, 1, 1, 2])

# =========================
# SUMATORIAS
# =========================
n = len(x)

Sx = np.sum(x)
Sy = np.sum(y)
Sx2 = np.sum(x**2)
Sxy = np.sum(x*y)

# =========================
# SISTEMA LINEAL
# =========================
# [Sx2  Sx] [m] = [Sxy]
# [Sx   n ] [b] = [Sy ]

A = np.array([
    [Sx2, Sx],
    [Sx,  n ]
])

B = np.array([Sxy, Sy])

# Resolver sistema
coef = np.linalg.solve(A, B)

m, b = coef

print("Parámetros encontrados:")
print(f"m = {m}")
print(f"b = {b}")

# =========================
# FUNCION
# =========================
def f(x):
    return m*x + b

# =========================
# GRAFICA
# =========================
x_fit = np.linspace(min(x)-1, max(x)+1, 200)
y_fit = f(x_fit)

plt.scatter(x, y, color='red', label="Datos")
plt.plot(x_fit, y_fit, 'blue', label="Ajuste lineal")

plt.title("Ajuste lineal por mínimos cuadrados")
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# DATOS
# =========================
x = np.array([
    3.38, 0.35, 2.07, -0.45, -0.56, -1.06, -2.78, 3.08, -3.99, -5.00,
    -3.18, -1.97, 2.37, -1.67, 4.09, -4.60, 2.68, 2.78, -3.79, -1.87
])

y = np.array([
    15.65, -11.20, -4.00, 18.03, 7.94, 15.32, -4.40, 1.39, -11.92, -90.24,
    6.92, 28.35, -2.41, -5.47, 35.91, -55.53, 0.77, 4.79, -27.05, 2.85
])

# =========================
# VALIDACIÓN
# =========================
if len(x) != len(y):
    raise ValueError("x e y deben tener el mismo tamaño")

n = len(x)

# =========================
# SUMATORIAS
# =========================
Sx = np.sum(x)
Sx2 = np.sum(x**2)
Sx3 = np.sum(x**3)
Sx4 = np.sum(x**4)
Sx5 = np.sum(x**5)
Sx6 = np.sum(x**6)

Sy = np.sum(y)
Sxy = np.sum(x*y)
Sx2y = np.sum((x**2)*y)
Sx3y = np.sum((x**3)*y)

# =========================
# SISTEMA CÚBICO
# =========================
A3 = np.array([
    [Sx6, Sx5, Sx4, Sx3],
    [Sx5, Sx4, Sx3, Sx2],
    [Sx4, Sx3, Sx2, Sx ],
    [Sx3, Sx2, Sx,  n  ]
])

B3 = np.array([Sx3y, Sx2y, Sxy, Sy])

a3, b3, c3, d3 = np.linalg.pinv(A3) @ B3

def f3(x):
    return a3*x**3 + b3*x**2 + c3*x + d3


def resumen_errores(y_real, y_estimado):
    errores_absolutos = np.abs(y_real - y_estimado)
    return {
        "promedio": np.mean(errores_absolutos),
        "maximo": np.max(errores_absolutos),
        "minimo": np.min(errores_absolutos),
        "desviacion_estandar": np.std(errores_absolutos),
    }


resumen_cubico = resumen_errores(y, f3(x))

print("Parámetros cúbicos encontrados:")
print(f"a3 = {a3}")
print(f"b3 = {b3}")
print(f"c3 = {c3}")
print(f"d3 = {d3}")
print()
print("Resumen del error absoluto para el ajuste cúbico:")
print(f"Error promedio = {resumen_cubico['promedio']}")
print(f"Error máximo = {resumen_cubico['maximo']}")
print(f"Error mínimo = {resumen_cubico['minimo']}")
print(f"Desviación estándar = {resumen_cubico['desviacion_estandar']}")

# =========================
# AJUSTE LINEAL
# =========================
A1 = np.array([[Sx2, Sx],
               [Sx,  n]])

B1 = np.array([Sxy, Sy])

m, b1 = np.linalg.pinv(A1) @ B1

def f1(x):
    return m*x + b1

# =========================
# AJUSTE CUADRÁTICO
# =========================
A2 = np.array([
    [Sx4, Sx3, Sx2],
    [Sx3, Sx2, Sx ],
    [Sx2, Sx,  n  ]
])

B2 = np.array([Sx2y, Sxy, Sy])

a2, b2, c2 = np.linalg.pinv(A2) @ B2

def f2(x):
    return a2*x**2 + b2*x + c2

# =========================
# GRAFICAR
# =========================
x_fit = np.linspace(min(x)-2, max(x)+2, 400)

plt.figure(figsize=(10,6))
plt.scatter(x, y, color='red', label="Datos", zorder=3)
plt.plot(x_fit, f1(x_fit), 'green', label="Lineal")
plt.plot(x_fit, f2(x_fit), 'blue', label="Cuadrático")
plt.plot(x_fit, f3(x_fit), 'purple', label="Cúbico")
plt.title("Comparación: Lineal vs Cuadrático vs Cúbico")
plt.legend()
plt.grid(True)
plt.show()

# =========================
# GRÁFICA DE DESVIACIÓN ESTÁNDAR RESPECTO A LA RECTA
# =========================
residuos_lineales = y - f1(x)
desviacion_lineal = np.std(residuos_lineales)

print()
print("Resumen del error respecto a la recta:")
print(f"Desviación estándar de los residuales = {desviacion_lineal}")

plt.figure(figsize=(10,6))
plt.scatter(x, y, color='red', label="Datos", zorder=3)
plt.plot(x_fit, f1(x_fit), color='green', linewidth=2, label="Recta ajustada")
plt.fill_between(
    x_fit,
    f1(x_fit) - desviacion_lineal,
    f1(x_fit) + desviacion_lineal,
    color='green',
    alpha=0.15,
    label="±1 desviación estándar"
)

for xi, yi in zip(x, y):
    yi_recta = f1(xi)
    plt.plot([xi, xi], [yi_recta, yi], color='gray', alpha=0.45, linewidth=1)

plt.title("Desviación estándar de los puntos respecto a la recta")
plt.legend()
plt.grid(True)
plt.show()
```
