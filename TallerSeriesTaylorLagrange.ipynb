```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

# Intervalos
x = np.linspace(-4, 4, 1000)

# Función
y = np.cos(x)

# Polinomios de Taylor
P0 = np.ones_like(x)

P2 = 1 - x**2/factorial(2)

P4 = P2 + x**4/factorial(4)


plt.figure(figsize=(10,6))
plt.plot(x, y, label='cos(x)', linewidth=2)

plt.plot(x, P0, label='Taylor orden 0')
plt.plot(x, P2, label='Taylor orden 2')
plt.plot(x, P4, label='Taylor orden 4')

plt.title('Aproximaciones de Taylor para cos(x) en x0=0')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Intervalos
x = np.linspace(-0.9, 0.9, 1000)

# Función
y = 1/(1-x)

# Polinomios de Taylor
P0 = np.ones_like(x)

P1 = 1 + x

P2 = 1 + x + x**2

P3 = 1 + x + x**2 + x**3

# Gráfica
plt.figure(figsize=(10,6))

plt.plot(x, y, label='1/(1-x)', linewidth=2)

plt.plot(x, P0, label='Taylor orden 0')
plt.plot(x, P1, label='Taylor orden 1')
plt.plot(x, P2, label='Taylor orden 2')
plt.plot(x, P3, label='Taylor orden 3')

plt.title('Aproximaciones de Taylor de 1/(1-x) en x0=0')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Intervalo alrededor de x0 = 1
x = np.linspace(0.2, 2.0, 1000)

# Función
y = np.log(x)

# Polinomios de Taylor
P1 = (x - 1)

P2 = P1 - (x - 1)**2 / 2

P3 = P2 + (x - 1)**3 / 3

P4 = P3 - (x - 1)**4 / 4

# Gráfica
plt.figure(figsize=(10,6))

plt.plot(x, y, label='ln(x)', linewidth=2)

plt.plot(x, P1, label='Taylor orden 1')
plt.plot(x, P2, label='Taylor orden 2')
plt.plot(x, P3, label='Taylor orden 3')
plt.plot(x, P4, label='Taylor orden 4')

plt.axvline(x=1, linestyle='--', label='x0 = 1')

plt.title('Aproximaciones de Taylor de ln(x) alrededor de x0 = 1')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Datos
x = np.array([0, 30, 60, 90])
y = np.array([0, 0.5, np.sqrt(3)/2, 1])

# Polinomio de Lagrange
P = lagrange(x, y)

print("Polinomio:")
print(P)

# Puntos para la grafica
xx = np.linspace(0, 90, 500)
yy = P(xx)

# Gráfica
plt.figure(figsize=(10,6))
plt.plot(xx, yy, label='Polinomio de Lagrange')
plt.scatter(x, y, s=60, label='Datos originales')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolación de Lagrange')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([1, 2, 3])
y = np.array([1, 2, 2])

# Polinomio obtenido
xx = np.linspace(0, 4, 500)
yy = -0.5*xx**2 + 2.5*xx - 1

# Gráfica
plt.figure(figsize=(8,5))
plt.plot(xx, yy, label='P(x) = -0.5x² + 2.5x - 1')
plt.scatter(x, y, s=60, label='Puntos dados')

plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de Lagrange')
plt.legend()
plt.show()
```
