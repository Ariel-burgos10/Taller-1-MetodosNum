## GRAFICA


```python
import numpy as np
xs = list (range(1,3))
xs = np.linspace(1, 3, 100)
def f(x):
    _y=x**3-3*x**2+x-1

    return _y
```

```python
ys = [f(x) for x in xs]
```

```python
import matplotlib.pyplot as plt


plt.plot(xs, ys,)
plt.xlabel("xs")
plt.ylabel("ys")
plt.title("Gráfica de ys vs xs")
plt.grid(True)
plt.show()


```
