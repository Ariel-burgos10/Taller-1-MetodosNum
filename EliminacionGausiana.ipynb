5. Dado el sistema lineal:
   
𝑥1 − 𝑥2 + 𝛼𝑥3 = −2,

−𝑥1 + 2𝑥2 − 𝛼𝑥3 = 3,

𝛼𝑥1 + 𝑥2 + 𝑥3 = 2.

b. Encuentre el valor(es) de 𝛼 para los que el sistema tiene un número infinito de soluciones.

c. Suponga que existe una única solución para una a determinada, encuentre la solución.

# Sistema lineal con parámetro \(\alpha\)

\[
\begin{cases}
x_1-x_2+\alpha x_3=-2\\
-x_1+2x_2-\alpha x_3=3\\
\alpha x_1+x_2+x_3=2
\end{cases}
\]

La matriz de coeficientes es:

\[
A=
\begin{pmatrix}
1 & -1 & \alpha\\
-1 & 2 & -\alpha\\
\alpha & 1 & 1
\end{pmatrix}
\]

Calculamos el determinante:

\[
\det(A)=-(\alpha-1)(\alpha+1)
\]

Entonces:

\[
\det(A)=0
\]

cuando:

\[
\alpha=1 \quad \text{o} \quad \alpha=-1
\]

---

## a. Valores de \(\alpha\) para los que no tiene solución

Para:

\[
\alpha=1
\]

el sistema se vuelve incompatible.

\[
\boxed{\alpha=1}
\]

Por lo tanto, para \(\alpha=1\), el sistema **no tiene solución**.

---

## b. Valores de \(\alpha\) para los que tiene infinitas soluciones

Para:

\[
\alpha=-1
\]

el sistema queda dependiente y compatible.

\[
\boxed{\alpha=-1}
\]

Por lo tanto, para \(\alpha=-1\), el sistema tiene **infinitas soluciones**.

---

## c. Solución única

Existe solución única cuando:

\[
\alpha \neq 1
\]

y

\[
\alpha \neq -1
\]

Es decir:

\[
\alpha \neq \pm 1
\]

La solución general es:

\[
\boxed{x_1=\frac{1}{\alpha-1}}
\]

\[
\boxed{x_2=1}
\]

\[
\boxed{x_3=-\frac{1}{\alpha-1}}
\]

---

# Resumen final

\[
\boxed{\alpha=1 \Rightarrow \text{no tiene solución}}
\]

\[
\boxed{\alpha=-1 \Rightarrow \text{infinitas soluciones}}
\]

\[
\boxed{\alpha \neq \pm 1 \Rightarrow \text{solución única}}
\]

\[
\boxed{
(x_1,x_2,x_3)=
\left(
\frac{1}{\alpha-1},
1,
-\frac{1}{\alpha-1}
\right)
}
\]
