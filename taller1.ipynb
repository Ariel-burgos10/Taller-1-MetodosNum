* Necesito: tokens, precio

tokens lista 0 - 1M

```python
import numpy as np 
```

```python
tokens = list(range(1_000_000))
```

```python
print(tokens[:10])
print(tokens[-5:])
```

```python

precio = []
for t in tokens:
    if t <= 272_000:
        precio.append(15 * t + 8)
    else:
        precio.append(10 * t - 4)
```

```python
import matplotlib.pyplot as plt

plt.plot(tokens, precio)

plt.xlabel("Tokens")
plt.ylabel("Precio")
plt.title("Precio vs Tokens")

plt.grid(True)
plt.show()
```
