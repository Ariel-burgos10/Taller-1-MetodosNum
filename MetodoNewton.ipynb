```python
import numpy as np

xs = list(range(-2, 6))
xs = list(np.linspace(-1, 5, 100))
```

```python
def f(x):
    _y = x**3 - 3 * x**2 + x - 1
    if np.isscalar(x):
        print(f"{x:.2f} \t {_y:.2f}")
    return _y


def df(x):
    _dy = 3 * x**2 - 6 * x + 1

    return _dy
```

```python
ys = [f(x) for x in xs]
```

```python
import matplotlib.pyplot as plt


plt.plot(xs, ys,)
plt.xlabel("xs")
plt.ylabel("ys")
plt.title("Gráfica de ys vs xs")
plt.grid(True)
plt.show()


```

```python
import matplotlib.pyplot as plt

# -------- Primer caso: x0 = -10 --------
x_vals1 = []
x = -10

for i in range(20):
    x_vals1.append(x)
    x = x - f(x)/df(x)

iters1 = list(range(len(x_vals1)))


# -------- Segundo caso: x0 = -8 --------
x_vals2 = []
x = -8

for i in range(20):
    x_vals2.append(x)
    x = x - f(x)/df(x)

iters2 = list(range(len(x_vals2)))


plt.figure(figsize=(18,6))

plt.plot(iters1, x_vals1, marker='o', label='x0 = -10')
plt.plot(iters2, x_vals2, marker='s', label='x0 = -8')

# Etiquetas de puntos primer caso
for i, x in enumerate(x_vals1):
    plt.text(i, x, f"{x:.2f}", fontsize=8)

# Etiquetas de puntos segundo caso
for i, x in enumerate(x_vals2):
    plt.text(i, x, f"{x:.2f}", fontsize=8)

plt.title("Comparación Método de Newton")
plt.xlabel("Iteración")
plt.ylabel("xₙ")
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def dibujar_tangentes(funcion, derivada, puntos, x_min=None, x_max=None, n=400):
    puntos = list(puntos)
    if not puntos:
        raise ValueError("Se requiere al menos un punto para dibujar las tangentes.")

    if x_min is None:
        x_min = min(puntos) - 2
    if x_max is None:
        x_max = max(puntos) + 2

    x_plot = np.linspace(x_min, x_max, n)

    plt.figure(figsize=(10, 6))
    plt.plot(x_plot, funcion(x_plot), color="black", linewidth=2, label="f(x)")
    plt.axhline(0, color="gray", linewidth=0.8)

    for idx, x0 in enumerate(puntos, start=1):
        y0 = funcion(x0)
        m = derivada(x0)
        tangente = y0 + m * (x_plot - x0)
        plt.plot(x_plot, tangente, linestyle="--", linewidth=1.4, label=f"Tangente {idx}")
        plt.scatter([x0], [y0], color="crimson", s=35, zorder=3)

    plt.scatter([puntos[-1]], [funcion(puntos[-1])], color="royalblue", s=50, zorder=4, label="Última iteración")
    plt.title("Método de Newton con tangentes")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()


dibujar_tangentes(f, df, xs_iter)
```
